```python
from geopy.geocoders import Nominatim
```

```python
nom = Nominatim()

geo_df  =pd.read_csv("http://pythonhow.com/supermarkets.csv")

geo_df['geo_info'] = geo_df.ix[:,1]+", "+geo_df.ix[:,2]+", "+geo_df.ix[:,3]+", "+geo_df.ix[:,4]

geo_df['coodinate'] = geo_df['geo_info'].apply(nom.geocode)

geo_df
```

```python
geo_df.coodinate[0].latitude
```

```python
geo_df["Latitude"] = geo_df.coodinate.apply(lambda x:x.latitude if x != None else None )
geo_df["Longitude"] = geo_df.coodinate.apply(lambda x:x.longitude if x != None else None )
```
